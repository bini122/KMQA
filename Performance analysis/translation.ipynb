```python
data_path = '/content/drive/MyDrive/korquad/data/more_data/PsyQA_example.json'

import json
with open(data_path, "r", encoding="utf-8") as reader:
    input_data = json.load(reader)
```

```python
import pandas as pd
from tqdm import tqdm
data = pd.DataFrame(columns=['raw_url', 'context', 'raw_question', 'raw_answer', 'annotated_question', 'is_impossible'])
uid = 0
for i in tqdm(range(len(input_data))):
  try:
    data.loc[uid, 'raw_url'] = input_data[i]['questionID']
    data.loc[uid, 'context'] = input_data[i]['question'] + " " + input_data[i]['answers'][0]['answer_text']
    data.loc[uid, 'raw_question'] = input_data[i]['question']
    data.loc[uid, 'raw_answer'] = input_data[i]['answers'][0]['answer_text']
    data.loc[uid, 'annotated_question'] = input_data[i]['question']
    data.loc[uid, 'is_impossible'] = "O"
    uid += 1
  except:
    print(i)
```

```python
from googletrans import Translator
from tqdm import tqdm

translator = Translator()

lang = 'ko'
for idx in tqdm(range(len(data))):
  try:
    result = translator.translate(data.loc[idx, 'context'], lang).text
    data.loc[idx, 'context'] = result
  except:
    data.loc[idx, 'context'] = 'pass'

  try:
    result = translator.translate(data.loc[idx, 'raw_question'], lang).text
    data.loc[idx, 'raw_question'] = result
  except:
    data.loc[idx, 'raw_question'] = 'pass'
  
  try:
    result = translator.translate(data.loc[idx, 'raw_answer'], lang).text
    data.loc[idx, 'raw_answer'] = result
  except:
    data.loc[idx, 'raw_answer'] = 'pass'

  try:
    result = translator.translate(data.loc[idx, 'annotated_question'], lang).text
    data.loc[idx, 'annotated_question'] = result
  except:
    data.loc[idx, 'annotated_question'] = 'pass'
```

```python
data.to_csv('chinese2korean.csv', index=False)
```

```python
data_path = '/content/drive/MyDrive/korquad/data/more_data/mentalhealth.json'

import json
with open(data_path, "r", encoding="utf-8") as reader:
    input_data = json.load(reader)
```

```python
import pandas as pd
from tqdm import tqdm
mentalhealth = pd.DataFrame(columns=['raw_url', 'context', 'raw_question', 'raw_answer', 'annotated_question', 'is_impossible'])
uid = 0
for i in tqdm(range(len(input_data['intents']))):
    lists=input_data['intents'][i]['patterns']
    for j in range(len(lists)):
        mentalhealth.loc[uid, 'raw_url'] = input_data['intents'][i]['tag']
        mentalhealth.loc[uid, 'context'] = input_data['intents'][i]['patterns'][j] + " " + input_data['intents'][i]['responses'][0]
        mentalhealth.loc[uid, 'raw_question'] = input_data['intents'][i]['patterns'][j]
        mentalhealth.loc[uid, 'raw_answer'] = input_data['intents'][i]['responses'][0]
        mentalhealth.loc[uid, 'annotated_question'] = input_data['intents'][i]['patterns'][j]
        mentalhealth.loc[uid, 'is_impossible'] = "O"
        uid += 1
```

```python
from googletrans import Translator
from tqdm import tqdm

translator = Translator()

lang = 'ko'
for idx in tqdm(range(len(mentalhealth))):
  try:
    result = translator.translate(mentalhealth.loc[idx, 'context'], lang).text
    mentalhealth.loc[idx, 'context'] = result
  except:
    mentalhealth.loc[idx, 'context'] = 'pass'

  try:
    result = translator.translate(mentalhealth.loc[idx, 'raw_question'], lang).text
    mentalhealth.loc[idx, 'raw_question'] = result
  except:
    mentalhealth.loc[idx, 'raw_question'] = 'pass'
  
  try:
    result = translator.translate(mentalhealth.loc[idx, 'raw_answer'], lang).text
    mentalhealth.loc[idx, 'raw_answer'] = result
  except:
    mentalhealth.loc[idx, 'raw_answer'] = 'pass'

  try:
    result = translator.translate(mentalhealth.loc[idx, 'annotated_question'], lang).text
    mentalhealth.loc[idx, 'annotated_question'] = result
  except:
    mentalhealth.loc[idx, 'annotated_question'] = 'pass'
```

```python
mentalhealth.to_csv('english2korean.csv', index=False)
```
